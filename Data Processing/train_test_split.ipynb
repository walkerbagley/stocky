```python
# Imports
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
data = pd.read_csv('merged_datasets.csv')

data.head(5)
```

```python
# get data in sequence
data = data.sort_values('Date')
data = data.reset_index(drop=True)

data.head(5)
```

```python
print(data.dtypes)
```

```python
cols = ['Price', 'Open', 'High', 'Low', 'Change %']

for col in cols:
    for i, entry in enumerate(data[col]):
        if type(entry) == str:
            entry = entry.replace(',', '')
            entry = entry.replace('%', '')
        entry = float(entry)
        data.at[i, col] = entry

data.head(5)
```

```python
numTrain = int(len(data) * 0.8)
numTest = len(data) - numTrain
print(numTrain, numTest, len(data))
```

```python
training = data.iloc[0:numTrain]
print(len(training))

testing = data.iloc[numTrain:len(data)]
testing = testing.reset_index(drop=True)
print(len(testing))
```

```python
display(training)
```

```python
display(testing)
```

```python
# Apply scaling to each column in cols
for col in cols:
    train_scaler = MinMaxScaler()
    test_scaler = MinMaxScaler()
    training[col] = train_scaler.fit_transform(training[[col]])
    testing[col] = test_scaler.fit_transform(testing[[col]])
```

```python
display(training)
```

```python
display(testing)
```

```python
training.to_csv("normalized_training_data.csv", index=False)
testing.to_csv("normalized_testing_data.csv", index=False)
```
