# Run and Test
This is the code to run and test a pretrained custom LSTM model on the testing dataset

```python
# Imports
import pandas as pd
from lstm import customLSTM
```

```python
# Specify path and load test data
path = 'Price-5-200E-0_8195.keras'
test = pd.read_csv('normalized_testing_data.csv')
```

```python
# Load Specified model with test data
loadmodel = customLSTM(path=path)

# Test on the data and plot the results
loadmodel.test(test, verbose = 0)
loadmodel.plotResults()
```
